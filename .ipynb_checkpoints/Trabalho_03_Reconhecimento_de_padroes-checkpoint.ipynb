## I) QDA (using the complete Bayesian discriminant formulation).

```python
def qda(X, y, verbose=False):
    
    # 1- calculate the probability a priori [1x1]
    probs_p = fp.prob_priori(X)

    # 2- average vector per class [1x34]
    u = fp.mean_vector_medio_per_feature(X)

    # 3- covariancy matrix by class [34x34]
    mcov = []
    for i in X['c35'].unique():
        mcov.append(fp.covariance_matrix(X.query('c35=='+str(i))))

    # 4- Make lamdba correction A = A + xI in the covariancy matrix
    mcov_lambda = []
    for i in range(X['c35'].unique().shape[0]):
        mcov_lambda.append(fp.correct_lambda(mcov[i]))
    
    # 5- Calculate the inverse of the lambda regulated covariance matrix
    mcov_inv = []
    for i in range(X['c35'].unique().shape[0]):
        mcov_inv.append(fp.inverse_matrix(mcov_lambda[i]))

    # 6- Calculate the determinant of each regulated covariance matrix
    det = []
    for i in range(X['c35'].unique().shape[0]):
        det.append(fp.covariance_matrix_determinant(mcov_lambda[i]))    
    
    """
      5 - Test, now take an element from the test base and apply    
      m in: ln|det(mcov_classe_i)| + (elem_teste-vetor_medio_classe_i).T*
          mcov_classe_i_inv*(elem_teste-vetor_md_classe_i) - 2*ln(P_priori(classe_i)) 
    """
    y_pred = []
    for i in y.index.values:
        valores = []
        for j in range(X['c35'].unique().shape[0]):
            valores.append(fp.test_elem(y.iloc[i,:-1],u.iloc[j],mcov_inv[j],det[j],probs_p[j]))    
        y_pred.append(valores.index(min(valores))+1)
    
    df_m_conf = fp.confusion_matrix(y['c35'],y_pred)

    if verbose: plt.plot_confusion_matrix(df_m_conf)
        
    return rep.print_results(df_m_conf)
```

```python
X,y = fp.kfold_shuffle()
print(qda(X,y,verbose=True))
```

```python
qda_metrics = pd.DataFrame({})
for _ in range(5):
    X,y = fp.kfold_shuffle()
    qda_metrics = qda_metrics.append(qda(X, y, verbose=False), ignore_index=True)
print(f'Average metrics \n{qda_metrics.mean()}')
```

## II) Naive Bayes (considering that the attributes of the problem are uncorrelated)

```python
def naives_bayes(X, y, verbose=False):
    """
        This function calculte the Naive Bayes algorithm
        
        @ X - Input samples
        @ y - output samples
        @ verbose - if you want plot confusion matrix
        
        @ return accuracy, recall, precision and f1-score
    """
    
    
    # 1- calculate the probability a priori [1x1]
    probs_p = fp.prob_priori(X)

    # 2- average vector per class [1x34]
    u = fp.mean_vector_medio_per_feature(X)

    mcov = []
    for i in X['c35'].unique():
        mcov.append(fp.covariance_matrix(X.query('c35=='+str(i))))
    
    # 3- covariancy matrix by class [34x34] knowing that they are uncorrelated
    mcov_descorrelacionada = []
    for i in range(X['c35'].unique().shape[0]):
        mcov_descorrelacionada.append(fp.clean_covariance(mcov[i]))

    # 4- Make lamdba correction A = A + xI in the covariancy matrix
    mcov_lambda = []
    for i in range(X['c35'].unique().shape[0]):
        mcov_lambda.append(fp.correct_lambda(mcov_descorrelacionada[i]))

    # 5- Calculate the inverse of the lambda regulated covariance matrix
    mcov_inv = []
    for i in range(X['c35'].unique().shape[0]):
        mcov_inv.append(fp.inverse_matrix(mcov_lambda[i]))
    
    # 6- Calculate the determinant of each regulated covariance matrix
    det = []
    for i in range(X['c35'].unique().shape[0]):
        det.append(fp.covariance_matrix_determinant(mcov_lambda[i]))    
    
    """
      5 - Test, now take an element from the test base and apply    
        
        min: ln|det(mcov_classe_i)| + (elem_teste-vetor_medio_classe_i).T
              *mcov_classe_i_inv*(elem_teste-vetor_md_classe_i) - 2*ln(P_priori(classe_i)) 
    
    """ 
    y_pred = []
    for i in y.index.values:
        valores = []
        for j in range(X['c35'].unique().shape[0]):
            valores.append(fp.test_elem(y.iloc[i,:-1],u.iloc[j],mcov_inv[j],det[j],probs_p[j]))    
        y_pred.append(valores.index(min(valores))+1)
        
    df_m_conf = fp.confusion_matrix(y['c35'],y_pred)
    if verbose: plt.plot_confusion_matrix(df_m_conf)
    
    return rep.print_results(df_m_conf)
```

```python
X,y = fp.kfold_shuffle()
print(naives_bayes(X, y, verbose=True))
```

## III) LDA (considering that all classifiers have the same covariance matrix and the same probability a priori).

‚Ä¢ OBS: to calculate the aggregate covariance matrix, one must calculate the covariance matrices for each class, and calculate the average of these.

- For each classifier, using the ‚Äúdermatology‚Äù database, perform the following experiment:

     + Use k-fold cross validation (k = 5) to generate the results.
     + For each of the 5 data partitions (between training and testing) perform the following procedure:
         + From the training data, calculate the mean and standard deviation of each attribute and normalize the z-score of the data (both training data and test data).
         + Calculate the necessary statistics (covariance matrix, average vector, a priori probability ...), for the classifier, from the training data.
         + Sort the test data.

```python
def lda(X, y, verbose=False):
    """
        This function calculte the Linear Discriminant Analysis
        
        @ X - Input samples
        @ y - output samples
        @ verbose - if you want plot confusion matrix
        
        @ return accuracy, recall, precision and f1-score
    
    """
    
    # 1- average vector per class [1x34]
    u= []
    for i in range(6):
        u.append(fp.average_vector(X.query('c35=='+str(i+1))))
    
    # 2- mcov_medio or aggregate is [34x34] add all the mcov of n classes and make the average mcov / n classes [34x34]
    mcovs = 0
    for i in X['c35'].unique():
        mcovs  += fp.covariance_matrix(X.query('c35=='+str(i)))
    mcov_medio = mcovs/6
    
    # 3- Make lamdba correction A = A + xI in the covariancy matrix
    mcov_medio_lambda = fp.correct_lambda(mcov_medio)
 
    # 4- Calculate the inverse of the regulated lambda covariance matrix
    mcov_medio_lambda_inv_d = fp.inverse_matrix(mcov_medio_lambda)
    
    """
      5 - Test, now take an element from the test base and apply    
         min: (elem_teste-vetor_medio_classe_i).T*mcov_classe_i_inv_media(agredada)*(elem_teste-vetor_md_classe_i)
    """
    
    y_pred = []
    for i in y.index.values:
        valores = []
        for j in range(len(u)):
            valores.append(fp.test_elem_lda(y.iloc[i,:-1],u[j],mcov_medio_lambda_inv_d))    
        y_pred.append(valores.index(min(valores))+1)
        
    df_m_conf = fp.confusion_matrix(y['c35'],y_pred)
    if verbose: plt.plot_confusion_matrix(df_m_conf)
    
    return rep.print_results(df_m_conf)
```

```python
X,y = fp.kfold_shuffle()
print(lda(X,y,verbose=True))
```

```python
lda_metrics = pd.DataFrame({})
for _ in range(5):
    X, y = fp.kfold_shuffle()
    lda_metrics = lda_metrics.append(lda(X, y, verbose=False), ignore_index=True)
print(f'Metrics average \n{lda_metrics.mean()}')
```

## 2.3) LDA (CDA) as a linear transformation.
- From the dermatology database, perform the following experiment:
    + Separate the data between training (70%) and testing (30%).
    + From the training data, calculate the mean and standard deviation of each attribute and normalize the z-score of the data (both training data and test data).
    + From the training data, apply the LDA method using the multiple projection approach (also known as Canonical Discriminant Analysis, CDA).
    + From the calculated eigenvalues, define the number of attributes necessary to carry out the data projection. (Note: maximum quantity = "number of classes - 1")
    + Define the W projection matrix.
    + Transform the training data from the W matrix.
    + Use the transformed data to train a Bayesian classifier (of your choice)
    + Use the projection matrix to transform the test data.
    + Sort the test data and get the classifier hit rate.

```python
def CDA(X):    

    # 1 - Calculate the Sw matrix
    Sw = 0
    for i in range(6):
        Sw += fp.covariance_matrix(X.query('c35=='+str(i+1)))
        
    # 2 - Calculate the average vector of each class
    us = fp.mean_vector_medio_per_feature(X)

    # 3 - Calculate the mean of the average vectors
    u = us.mean()
    
    # 4 - Calculate the Sb projection matrix
    Sb = 0
    for i in X['c35'].unique():
        Sb  += fp.calculate_projection(X.query('c35=='+str(i)),u,us.iloc[i-1]).values
    
    # 5 - Calculate eigenvalues and eigenvectors
    m_aa_av = fp.inverse_matrix(Sw)*Sb
    
    a_values, a_vectors = np.linalg.eig(m_aa_av)
    df_a_values = pd.DataFrame(a_values)
    df_a_vectors = pd.DataFrame(a_vectors)
    
    # 6 - Search for the eigenvectors of the biggest eigenvalues
    W = pd.DataFrame({})
    for i in df_a_values.sort_values(by=0,ascending=True)[:5].index.values:
        W = W.append(df_a_vectors.iloc[i],ignore_index=True)

    return W
```

```python
naives_bayes_metrics = pd.DataFrame({})
for _ in range(5):
    X,y = fp.split_train_test_70_30(data_derma)
    W = CDA(X)
    Xtr_t =  fp.data_transform(W,X)
    Xts_t =  fp.data_transform(W,y)
    naives_bayes_metrics = naives_bayes_metrics.append(naives_bayes(Xtr_t,Xts_t,verbose=False),ignore_index=True)
print(f'Metrics average \n{naive_bayes_metrics.mean()}')
```
